### **Question 29 : synth√®se de traitement (Correction logique) (2.0 pts)**
La fonction `analyser_phrase` doit retourner un dictionnaire contenant une liste de `mots_nettoy√©s` (`['le', 'tal', 'est', 'cool']`) et le `nb_caracteres` **total de la phrase originale** (avec espaces et ponctuation). Corrigez les erreurs de calcul (`nb_caracteres` et `mots_nettoy√©s`).

````python
phrase = "Le TAL est cool."
ponctuation_a_enlever = ".", ",", "!", "?"

def nettoyer_mot_simple(mot):
    # La fonction est correcte ici si la ponctuation est un tuple de 4 √©l√©ments.
    mot = mot.lower()
    return mot.strip(ponctuation_a_enlever)

def analyser_phrase(phrase):
    mots = phrase.split()
    mots_nettoy√©s = [nettoyer_mot_simple(mot) for mot in mots]

    nb_caracteres = len(mots)

    # Assurez-vous qu'il n'y a pas de cha√Æne vide si le split() a laiss√© un espace.
    mots_nettoy√©s = [mot for mot in mots_nettoy√©s if mot]

    return {
        'mots_nettoy√©s': mots_nettoy√©s,
        'nb_caracteres': nb_caracteres
    }

valeur_Q29 = analyser_phrase(phrase)
print(f"R√©sultat Q29 : {valeur_Q29}")
````
