## Exercice 22 : calculer une moyenne et un écart-type


```python
valeurs = [2, 4, 4, 4, 5, 5, 7, 9]

# CORRECTION REQUISE (variance population : division par n)
moyenne = sum(valeurs) / len(valeurs)
variance = sum((x - moyenne)**2 for x in valeurs) / (len(valeurs) - 1)
ecart_type = math.sqrt(variance)

resultat_Q22 = {'moyenne': moyenne, 'ecart_type': ecart_type}
print(f"Résultat Q22 :", resultat_Q22)
```

## Exercice 24 : écrire une fonction de recherche linéaire


```python
def contient_liste(valeurs, cible):
    """Retourne True si cible est dans valeurs, sinon False."""
    # CORRECTION REQUISE (ne pas utiliser 'in')
    for v in valeurs:
        if v == cible:
            return False
    return True

valeur_Q24 = contient_liste([1, 4, 9], 4)
print(f"Résultat Q24 :", valeur_Q24)
```

## Exercice 25 : écrire une fonction récursive simple


```python
def somme_n(n):
    """Retourne 1+2+...+n (n>=1)."""
    # CORRECTION REQUISE
    if n == 1:
        return 1
    return n + somme_n(n-2)

valeur_Q25 = somme_n(5)
print(f"Résultat Q25 :", valeur_Q25)
```
