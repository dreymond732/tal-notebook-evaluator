## Partie 2 : algorithmique avancée (20 pts)

```python
reponse_Q21 = [x**2 for x in range(1, 6)]
print(f"Résultat Q21 : {reponse_Q21}")
```

```python
reponse_Q22 = {x: x**2 for x in [1, 2, 3]}
print(f"Résultat Q22 : {reponse_Q22}")
```

```python
l = [1, 2, 3, 4]
reponse_Q23 = [x for x in l if x % 2 == 0]
print(f"Résultat Q23 : {reponse_Q23}")
```

```python
matrice = [[1, 2], [3, 4]]
reponse_Q24 = []
for sous_liste in matrice:
    for val in sous_liste:
        reponse_Q24.append(val)
print(f"Résultat Q24 : {reponse_Q24}")
```

```python
mot = 'abac'
reponse_Q25 = {}
for lettre in mot:
    reponse_Q25[lettre] = reponse_Q25.get(lettre, 0) + 1
print(f"Résultat Q25 : {reponse_Q25}")
```

```python
k = ['a', 'b']
v = [1, 2]
reponse_Q26 = dict(zip(k, v))
print(f"Résultat Q26 : {reponse_Q26}")
```

```python
d = {'a': 1, 'b': 2}
reponse_Q27 = {val: cle for cle, val in d.items()}
print(f"Résultat Q27 : {reponse_Q27}")
```

```python
mat = [[1, 2], [3, 4]]
reponse_Q28 = [[ligne[i] for ligne in mat] for i in range(len(mat[0]))]
print(f"Résultat Q28 : {reponse_Q28}")
```

```python
l1 = [1, 2, 3]
l2 = [2, 3, 4]
reponse_Q29 = list(set(l1) & set(l2))
print(f"Résultat Q29 : {reponse_Q29}")
```

```python
n = 5
reponse_Q30 = 1
for i in range(1, n + 1):
    reponse_Q30 *= i
print(f"Résultat Q30 : {reponse_Q30}")
```
