### **Question 8 : filtrage conditionnel (2.0 pts)**
Cr√©ez une liste `adultes` contenant seulement les noms des personnes ayant 18 ans ou plus : `['Alice', 'Charlie']`.

```python
personnes = {'Alice': 25, 'Bob': 17, 'Charlie': 30}
adultes = []

# CORRECTION REQUISE
for nom, age in personnes.items():
    if age < 18: # Erreur de condition
        adultes.append(nom)

adultes.sort() # Pour l'ordre de v√©rification
valeur_Q8 = adultes
print(f"R√©sultat Q8 : {valeur_Q8}")
```
