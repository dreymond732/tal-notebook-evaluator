## Partie 1 : Échauffement et compréhensions

```python
# Exercice 1 : Carrés des pairs (0 à 20 inclus)
# CORRECTION : Compréhension avec if

reponse_Q1 = [x**2 for x in range(21) if x % 2 == 0]

print("Résultat Q1 :", reponse_Q1)
```

```python
# Exercice 2 : Filtrage de mots (commençant par voyelle)
mots = ["chat", "arbre", "python", "ecole", "igloo", "banane", "orange"]
voyelles = {'a', 'e', 'i', 'o', 'u', 'y'}

# CORRECTION
reponse_Q2 = [m for m in mots if m[0].lower() in voyelles]

print("Résultat Q2 :", reponse_Q2)
```

```python
# Exercice 3 : Dictionnaire de longueurs
phrase = "le traitement automatique du langage est passionnant"

# CORRECTION : Dict comprehension sur split()
reponse_Q3 = {mot: len(mot) for mot in phrase.split()}

print("Résultat Q3 :", reponse_Q3)
```

```python
# Exercice 4 : Inversion de phrase
# CORRECTION : Split, slicing inverse, join

mots = phrase.split()
reponse_Q4 = " ".join(mots[::-1])

print("Résultat Q4 :", reponse_Q4)
```

```python
# Exercice 5 : Association (Zip)
cles = ["nom", "prenom", "age"]
valeurs = ["Dupont", "Jean", 25]

# CORRECTION : dict(zip(...))
reponse_Q5 = dict(zip(cles, valeurs))

print("Résultat Q5 :", reponse_Q5)
```

## Partie 2 : Manipulation de chaînes et logique

```python
# Exercice 6 : Palindrome
mot_test = "radar"

# CORRECTION
reponse_Q6 = mot_test == mot_test[::-1]

print("Résultat Q6 :", reponse_Q6)
```

```python
# Exercice 7 : Anagrammes
mot1 = "chien"
mot2 = "niche"

# CORRECTION : sorted() == sorted()
reponse_Q7 = sorted(mot1) == sorted(mot2)

print("Résultat Q7 :", reponse_Q7)
```

```python
# Exercice 8 : Distance de Hamming
adn1 = "AGCTTAG"
adn2 = "AGCTATG"

# CORRECTION : Compter les différences position par position
reponse_Q8 = sum(1 for a, b in zip(adn1, adn2) if a != b)

print("Résultat Q8 :", reponse_Q8)
```

```python
# Exercice 9 : Générateur d'acronyme
organisation = "Traitement Automatique des Langues"

# CORRECTION
reponse_Q9 = "".join([w[0].upper() for w in organisation.split()])

print("Résultat Q9 :", reponse_Q9)
```

```python
# Exercice 10 : Chiffrement de César
message = "abc"
decalage = 1

# CORRECTION
reponse_Q10 = "".join([chr(ord(c) + decalage) for c in message])

print("Résultat Q10 :", reponse_Q10)
```

## Partie 3 : Structures de données complexes

```python
# Exercice 11 : Aplatissement de liste
liste_imbriquee = [[1, 2], [3, 4], [5, 6]]

# CORRECTION : Double boucle dans compréhension
reponse_Q11 = [item for sublist in liste_imbriquee for item in sublist]

print("Résultat Q11 :", reponse_Q11)
```

```python
# Exercice 12 : Fusion de dictionnaires (somme)
d1 = {'a': 10, 'b': 20, 'c': 30}
d2 = {'b': 5, 'c': 10, 'd': 40}

# CORRECTION
reponse_Q12 = d1.copy()
for k, v in d2.items():
    reponse_Q12[k] = reponse_Q12.get(k, 0) + v

print("Résultat Q12 :", reponse_Q12)
```

```python
# Exercice 13 : Inversion de dictionnaire
cours = {'Alice': 'Maths', 'Bob': 'Maths', 'Charlie': 'Physique'}

# CORRECTION
reponse_Q13 = {}
for etudiant, matiere in cours.items():
    if matiere not in reponse_Q13:
        reponse_Q13[matiere] = []
    reponse_Q13[matiere].append(etudiant)

# Note: le correcteur triera si besoin, mais ici l'ordre d'insertion est souvent préservé
print("Résultat Q13 :", reponse_Q13)
```

```python
# Exercice 14 : Tri complexe (par score décroissant)
scores = [('Alice', 12), ('Bob', 15), ('Charlie', 10)]

# CORRECTION : lambda sur x[1], reverse=True
reponse_Q14 = sorted(scores, key=lambda x: x[1], reverse=True)

print("Résultat Q14 :", reponse_Q14)
```

```python
# Exercice 15 : Élément le plus fréquent
data = [1, 2, 3, 2, 4, 2, 5, 1]

# CORRECTION : max avec key=data.count ou Counter
reponse_Q15 = max(set(data), key=data.count)

print("Résultat Q15 :", reponse_Q15)
```

## Partie 4 : Algorithmes pour le TAL

```python
# Exercice 16 : Tokenisation simple
raw_text = "Bonjour, comment ça va?"

# CORRECTION
reponse_Q16 = raw_text.replace(',', ' ').replace('?', ' ').lower().split()

print("Résultat Q16 :", reponse_Q16)
```

```python
# Exercice 17 : Suppression des mots vides (Stopwords)
tokens = ['le', 'chat', 'est', 'noir', 'et', 'le', 'chien', 'dort']
stopwords = {'le', 'est', 'et'}

# CORRECTION
reponse_Q17 = [t for t in tokens if t not in stopwords]

print("Résultat Q17 :", reponse_Q17)
```

```python
# Exercice 18 : Bigrammes
mots = ['je', 'suis', 'content']

# CORRECTION
reponse_Q18 = [(mots[i], mots[i+1]) for i in range(len(mots)-1)]

print("Résultat Q18 :", reponse_Q18)
```

```python
# Exercice 19 : Trigrammes
mots = ['un', 'deux', 'trois', 'quatre']

# CORRECTION
reponse_Q19 = [(mots[i], mots[i+1], mots[i+2]) for i in range(len(mots)-2)]

print("Résultat Q19 :", reponse_Q19)
```

```python
# Exercice 20 : Indice de Jaccard
set_A = {'chat', 'chien', 'souris'}
set_B = {'chat', 'oiseau', 'souris'}

# CORRECTION : |A inter B| / |A union B|
inter = len(set_A & set_B)
union = len(set_A | set_B)
reponse_Q20 = inter / union

print("Résultat Q20 :", reponse_Q20)
```

## Partie 5 : Avancé / Vectorisation

```python
# Exercice 21 : Calcul de TF (Term Frequency)
doc = ['a', 'b', 'a', 'c', 'a', 'b']

# CORRECTION : Count 'a' / Total
reponse_Q21 = doc.count('a') / len(doc)

print("Résultat Q21 :", reponse_Q21)
```

```python
# Exercice 22 : Produit scalaire (Dot product)
vec1 = [1, 2, 3]
vec2 = [4, 5, 6]

# CORRECTION : sum(x*y)
reponse_Q22 = sum(a*b for a, b in zip(vec1, vec2))

print("Résultat Q22 :", reponse_Q22)
```

```python
# Exercice 23 : Plus long préfixe commun
mots = ["programmation", "programme", "progrès"]

# CORRECTION : Approche simple avec os.path.commonprefix ou boucle
import os
reponse_Q23 = os.path.commonprefix(mots)

print("Résultat Q23 :", reponse_Q23)
```

```python
# Exercice 24 : Matrice de co-occurrence (simplifiée)
vocab = ["chat", "chien"]
contexte = ["le", "chat", "mange", "le", "chien"]

# CORRECTION
reponse_Q24 = {mot: contexte.count(mot) for mot in vocab}

print("Résultat Q24 :", reponse_Q24)
```

```python
# Exercice 25 : Pipeline complet
texte_brut = "  Python est Super!!  "

# CORRECTION
reponse_Q25 = texte_brut.strip().lower().replace('!', '').split()

print("Résultat Q25 :", reponse_Q25)
```
