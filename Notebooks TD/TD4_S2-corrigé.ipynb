## Partie 1 : Bases et UnicitÃ©

```python
# Q1 : DÃ©doublonnage
animaux_liste = ['chat', 'chien', 'chat', 'oiseau', 'chien']

# CORRECTION : set() transforme la liste en ensemble et Ã©limine les doublons
animaux_set = set(animaux_liste)

valeur_Q1 = animaux_set
print(f"RÃ©sultat Q1 : {valeur_Q1}")
```

```python
# Q2 : Modification (Ajout)
rongeurs = {'chat', 'chien'}

# CORRECTION : .add() pour un Ã©lÃ©ment unique
rongeurs.add('souris')

valeur_Q2 = rongeurs
print(f"RÃ©sultat Q2 : {valeur_Q2}")
```

```python
# Q3 : Appartenance
vocabulaire = {'corpus', 'token', 'lemme'}

# CORRECTION : mot-clÃ© 'in'
est_present = 'token' in vocabulaire

valeur_Q3 = est_present
print(f"RÃ©sultat Q3 : {valeur_Q3}")
```

## Partie 2 : OpÃ©rations Ensemblistes

```python
# Q4 : Intersection (&)
doc_A = {'python', 'code', 'data'}
doc_B = {'java', 'code', 'data'}

# CORRECTION
commun = doc_A & doc_B

valeur_Q4 = commun
print(f"RÃ©sultat Q4 : {valeur_Q4}")
```

```python
# Q5 : Union (|)
set_A = {'python', 'algo'}
set_B = {'data', 'code'}

# CORRECTION
tout = set_A | set_B

valeur_Q5 = tout
print(f"RÃ©sultat Q5 : {valeur_Q5}")
```

```python
# Q6 : DiffÃ©rence (-)
fruits = {'pomme', 'orange', 'citron', 'poire'}
agrumes = {'orange', 'citron'}

# CORRECTION
reste = fruits - agrumes

valeur_Q6 = reste
print(f"RÃ©sultat Q6 : {valeur_Q6}")
```

```python
# Q7 : DiffÃ©rence symÃ©trique (^)
os_serveur = {'linux', 'macos'}
os_client = {'windows', 'macos'}

# CORRECTION
exclusifs = os_serveur ^ os_client

valeur_Q7 = exclusifs
print(f"RÃ©sultat Q7 : {valeur_Q7}")
```

```python
# Q8 : Sous-ensemble
dictionnaire = {'le', 'chat', 'dort', 'mange'}
phrase = {'le', 'chat'}

# CORRECTION
est_inclus = phrase.issubset(dictionnaire)

valeur_Q8 = est_inclus
print(f"RÃ©sultat Q8 : {valeur_Q8}")
```

## Partie 3 : Simplification Algorithmique

```python
# Q9 : Optimisation recherche O(1)
liste_utilisateurs = ['user1', 'user2', 'user3', 'user4']
recherche = 'user3'

# CORRECTION
base_donnees = set(liste_utilisateurs)
est_trouve = recherche in base_donnees

valeur_Q9 = est_trouve
print(f"RÃ©sultat Q9 : {valeur_Q9}")
```

```python
# Q10 : Nettoyage donnÃ©es
tags_sales = ['tag1', 'tag2', 'tag1', 'tag3', 'tag2']

# CORRECTION
tags_uniques = set(tags_sales)

valeur_Q10 = tags_uniques
print(f"RÃ©sultat Q10 : {valeur_Q10}")
```

```python
# Q11 : Intersection listes
phrase1 = ['le', 'chat', 'est', 'noir']
phrase2 = ['le', 'chien', 'est', 'blanc']

# CORRECTION
mots_communs = set(phrase1) & set(phrase2)

valeur_Q11 = mots_communs
print(f"RÃ©sultat Q11 : {valeur_Q11}")
```

```python
# Q12 : Gestion absents (DiffÃ©rence)
classe_entiere = {'Jean', 'Marie', 'Paul', 'Sophie'}
devoirs_rendus = {'Marie', 'Sophie'}

# CORRECTION
absents = classe_entiere - devoirs_rendus

valeur_Q12 = absents
print(f"RÃ©sultat Q12 : {valeur_Q12}")
```

```python
# Q13 : Jaccard
set1 = {'pomme', 'banane', 'cerise'}
set2 = {'pomme', 'banane', 'fraise'}

# CORRECTION
intersection = set1 & set2 # {'pomme', 'banane'} (taille 2)
union = set1 | set2        # {'pomme', 'banane', 'cerise', 'fraise'} (taille 4)
jaccard = len(intersection) / len(union)

valeur_Q13 = jaccard
print(f"RÃ©sultat Q13 : {valeur_Q13}")
```

```python
# Q14 : Pangramme (Superset)
alphabet = set("abcdefghijklmnopqrstuvwxyz")
phrase = "Portez ce vieux whisky au juge blond".lower()

# CORRECTION
lettres_phrase = set(phrase)
# On vÃ©rifie si lettres_phrase contient tout l'alphabet (>= est l'opÃ©rateur pour superset)
est_pangramme = lettres_phrase >= alphabet 

valeur_Q14 = est_pangramme
print(f"RÃ©sultat Q14 : {valeur_Q14}")
```
