```python
personnes = {'Alice': 25, 'Bob': 17, 'Charlie': 30}
adultes = []

# CORRECTION REQUISE
for nom, age in personnes.items():
    if age >= 18:
        adultes.append(nom)

adultes.sort()
valeur_Q8 = adultes
print(f"RÃ©sultat Q8 : {valeur_Q8}")
```

```python
texte = "baba"
freq = {}

# CORRECTION REQUISE
for lettre in texte:
    if lettre in freq:
        freq[lettre] += 1
    else:
        freq[lettre] = 1

valeur_Q9 = freq
print(f"RÃ©sultat Q9 : {valeur_Q9}")
```

## Partie 3 : algorithmique avancÃ©e (9 pts)

```python
matrice = [[1, 2], [3, 4], [5, 6]]
liste_plate = []

# CORRECTION REQUISE
for sous_liste in matrice:
    for element in sous_liste:
        liste_plate.append(element)

valeur_Q10 = liste_plate
print(f"RÃ©sultat Q10 : {valeur_Q10}")
```

```python
fib = [0, 1]

# CORRECTION REQUISE
while True:
    prochain = fib[-1] + fib[-2]
    if prochain >= 10:
        break
    fib.append(prochain)

valeur_Q11 = fib
print(f"RÃ©sultat Q11 : {valeur_Q11}")
```

```python
def est_anagramme(mot1, mot2):
    # CORRECTION REQUISE (Astuce : trier les lettres)
    return sorted(mot1) == sorted(mot2)

valeur_Q12 = est_anagramme("gare", "rage")
print(f"RÃ©sultat Q12 : {valeur_Q12}")
```

```python
nombres = [5, 8, 12, 15, 20, 3]

# CORRECTION REQUISE
resultat = [x for x in nombres if x % 2 == 0 and x > 10]

valeur_Q13 = resultat
print(f"RÃ©sultat Q13 : {valeur_Q13}")
```
