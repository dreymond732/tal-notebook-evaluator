## Exercice 13 : tri avancé

## Exercice 22 : statistiques

```python
valeurs = [2, 4, 4, 4, 5, 5, 7, 9]

# CORRECTION : Variance Population (division par N et non N-1)
moyenne = sum(valeurs) / len(valeurs)
variance = sum((x - moyenne)**2 for x in valeurs) / len(valeurs)
ecart_type = math.sqrt(variance)

resultat_Q22 = {'moyenne': moyenne, 'ecart_type': ecart_type}
print(f"Résultat Q22 : {resultat_Q22}")
```

## Exercice 24 : recherche linéaire

```python
def contient_liste(valeurs, cible):
    # CORRECTION : Return True si trouvé, False à la fin
    for v in valeurs:
        if v == cible:
            return True
    return False

valeur_Q24 = contient_liste([1, 4, 9], 4)
print(f"Résultat Q24 : {valeur_Q24}")
```

## Exercice 25 : récursivité

```python
def somme_n(n):
    if n == 1:
        return 1
    # CORRECTION : n-1 pour descendre vers 1
    return n + somme_n(n-1)

valeur_Q25 = somme_n(5)
print(f"Résultat Q25 : {valeur_Q25}")
```

## Exercice 28 : stopwords

```python
phrase = "le tal et la python"
stopwords = {"le", "la", "les"}
tokens = phrase.split()

# CORRECTION : 'not in' pour garder les mots utiles
tokens_filtres = [t for t in tokens if t not in stopwords]

print(f"Résultat Q28 : {tokens_filtres}")
```

## Exercice 29 : produit cartésien
